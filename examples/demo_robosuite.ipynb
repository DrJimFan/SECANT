```python
from secant.envs.robosuite import make_robosuite, ALL_TASKS
```

```python
ALL_TASKS
```

```python
env = make_robosuite(
    task="Door",
    mode="train",
    scene_id=0,
    image_height=168,
    image_width=168,
)
```

```python
env.reset()
num_step = 0
while True:
    action = env.action_space.sample()
    obs, reward, done, info = env.step(action)
    num_step += 1
    if done:
        print(f"The episode terminates after {num_step} steps.")
        break
```
